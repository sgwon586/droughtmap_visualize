```python
import pandas as pd

path = "C:\\Users\\김현우\\Downloads\\시군별_논밭별_경지면적(강원)_20251120210149.csv" #(파일경로)
df = pd.read_csv(path, encoding ='cp949')
# print(df) # 실험용 실제 코드 x
```

```python
print(df)

df.columns =["구분", "시군", "경지면적(계)", "논", "밭"]

df["경지면적(계)"] = pd.to_numeric(df["경지면적(계)"], errors="coerce")
df["논"] = pd.to_numeric(df["논"], errors="coerce")
df["밭"] = pd.to_numeric(df["밭"], errors="coerce")
```

```python
# 강원도 전체(소계) 제외한 18개 시만 추출
df_region = df[df["시군"].notna()]
df_region = df_region[df_region["시군"] != "소계"]
```

```python
# 농경지 면적 비율 계산 --> 이게 필요한 게 맞나?
df_region["논비율(%)"] = df_region["논"] / df_region["경지면적(계)"] * 100
df_region["밭비율(%)"] = df_region["밭"] / df_region["경지면적(계)"] * 100
```

```python
# 최종 결과 확인
print(df_region)
```
