```python
import pandas as pd

path1 = "C:\\Users\\home\\Documents\\GitHub\\DroughtMap\\DATA\\시군별_논밭별_경지면적(강원)_20251120210149.csv" #(파일경로)
path2 = "C:\\Users\\home\\Documents\\GitHub\\DroughtMap\\DATA\\강원도 시군별 면적.csv"

df_farmland = pd.read_csv(path1, encoding ='cp949')
df_area = pd.read_csv(path2, encoding ='cp949')

print(df_farmland)
print(df_area)
```

```python
# 헤더 줄 제거 (구분-합계 써져있는 줄)
# df_farmland = df_farmland.drop([0, 1, 2])
# df_area = df_area.drop([0, 1])

# 농경지 데이터 컬럼 정리
df_farmland.columns = ["시군", "경지면적(계)", "논", "밭"]

# 숫자로 변환
for col in ["경지면적(계)", "논", "밭"]:
    df_farmland[col] = pd.to_numeric(df_farmland[col], errors="coerce")

# "소계" 제거
df_farmland = df_farmland[df_farmland["시군"] != "소계"]

# 시군 면적 데이터 컬럼 정리
df_area.columns = ["시군", "전체면적", "비율(도내)"]
df_area["전체면적"] = pd.to_numeric(df_area["전체면적"], errors="coerce")

# 두 데이터 결합
df_merge = pd.merge(df_farmland, df_area, on="시군", how="inner")

# 전체 면적 대비 농경지 비율 계산
# (헥타르와 킬로제곱미터 수치 표준화로 *1000을 함(1헥타르 = 1/1000킬로제곱미터), *100은 비율 계산)
df_merge["농경지비율(%)"] = (df_merge["경지면적(계)"] * 1000 / df_merge["전체면적"] * 100) / 100000
## df_merge["농경지비율(%)"] = round(df_merge["경지면적(계)"] / df_merge["전체면적"] , 3)

# 출력
print(df_merge[["시군", "전체면적", "경지면적(계)", "농경지비율(%)"]])
```

```python
# C:\Users\김현우\Downloads 폴더에 csv로 저장
save_path = "C:\\Users\\home\\Documents\\GitHub\\DroughtMap\\DATA\\강원도_농경지_면적비율(최종).csv"
df_merge.to_csv(save_path, encoding='cp949', index=False)
```
